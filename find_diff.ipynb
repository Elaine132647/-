```python
file1_path = 'company_new.txt'
file2_path = 'record_comp_new.txt'

def read_file(file_path):
    """Read the content of a file and return it as a set of lines."""
    with open(file_path, 'r') as file:
        return set(line.strip() for line in file)

def find_difference(set1, set2):
    """Find items in set1 that are not in set2."""
    return set1 - set2

def main():
    set1 = read_file(file1_path)
    set2 = read_file(file2_path)

    difference = find_difference(set1, set2)

    print("Items in the first file but not in the second file:")
    record_path = 'different_new.txt'
    with open(record_path,'w') as f:
        for item in difference:
            f.write(item +'\n')
            print(item)

if __name__ == "__main__":
    main()
```
